### **Impact Analysis - "Income" and "Fitness" Attribute:**

```python
df["Income"].describe()

# Income<=40000 --> "<=40000"
# >40000 to 70000 --> "<=70000"
# 70000 to 120000 --> ">70000"
```

```python
df["Income_Category"] = df["Income"].apply(
    lambda income: "<=40000" if income <= 40000
    else "<=70000" if 40000 < income <= 70000
    else ">70000" )
```

```python
df.head()
```

--------

```python
df["Fitness"].describe()

# 1 and 2 --> "Level_1_2"
# 3 --> "Level_3"
# 4 and 5 --> "Level_4_5"
```

```python
df["Fitness_category"] = df["Fitness"].apply(lambda x: "Level_1_2" if x<3 else "Level_3" if x==3 else "Level_4_5")
```

```python
df.head()
```

--------

#### **"Income" Attribute:**

```python
pd.crosstab(index=df["Product"], columns=df["Income_Category"], margins=True, margins_name="Total_Count")
```

##### **"Marginal Probability":**

```python
pd.crosstab(index=df["Product"], columns=df["Income_Category"], margins=True, margins_name="Total_Count", normalize=True)
```

**Insight 11A:**

Overall, KP281 is highly preferrable product, than KP481 and KP481 is Highly Preferrable than product KP781.

- KP281 Product is preferrable by Two Income category. Order:- Income<=70000(High or top buyers) and Income<=40000 (Medium buyers). Income>70000 is not a buyer for this product.

- KP481 Product is preferrable by Two Income category. Order:- Income<=70000(High or top buyers) and Income<=40000 (very low buyers). Income>70000 is not a buyer for this product.

- KP781 Product is preferrable by Two Income category. Order:- Income>70000(High or top buyers) and Income<=70000 (Medium buyers). Income<=40000 is not a buyer for this product.

##### **"Conditional Probability":**

```python
pd.crosstab(index=df["Product"], columns=df["Income_Category"], margins=True, margins_name="Total_Count", normalize="columns")
```

**Insights 11B:**

- If Income >70000 is given, then Product KP781 should be highly preferrable. Rest two products( KP281 and KP481 ) are not preferrable.

- If Income is between >40000 and <=70000 given, then Product KP281 and KP481 are more preferrable than KP781.

- If Income <=40000 is given, then Product KP281 should be highly preferable than Product KP481. Rest One product KP781 is not preferrable.
